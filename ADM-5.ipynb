```python
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.probability import FreqDist
from wordcloud import WordCloud
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, ne_chunk
from nltk.tree import Tree


```

```python
text = """
Stanford University is located in California. It is a great place for learning and research. 
Artificial Intelligence and Natural Language Processing are major fields here.
"""

# 1. Sentence Tokenization
sentences = sent_tokenize(text)
print("\n1. Sentence Tokenization:\n", sentences)

```

```python
# 2. Word Tokenization
words = word_tokenize(text)
print("\n2. Word Tokenization:\n", words)

```

```python
# 3. Frequency Distribution of Words
fdist = FreqDist(words)
print("\n3. Frequency Distribution:")
fdist.plot(20, title="Word Frequency Distribution")

```

```python
# 4. Creating Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud")
plt.show()

```

```python
# 5. Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in words]
print("\n5. Stemming:\n", stemmed_words)
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
# 6. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print("\n6. Lemmatization:\n", lemmatized_words)

```

```python
nltk.download('averaged_perceptron_tagger')
# 7. POS Tagging
pos_tags = pos_tag(words)
print("\n7. POS Tagging:\n", pos_tags)
```

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
# 8. Named Entity Recognition (NER)
ner_tree = ne_chunk(pos_tags)
print("\n8. Named Entity Recognition:")
for subtree in ner_tree:
    if isinstance(subtree, Tree):
        print("Named Entity:", " ".join([token for token, pos in subtree.leaves()]), "Label:", subtree.label())
```
